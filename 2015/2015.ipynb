```python
# Data imports
from pathlib import Path

base_path = Path('/home/meek/github/gojo/advent_of_code/2015/data/')
if not base_path.exists():
    raise FileNotFoundError(f"Base path not found: {base_path}")

# Discover all .txt input files in the directory
inp_files = sorted([str(p) for p in base_path.iterdir() if p.is_file() and p.suffix == '.txt'])
print(f'Found {len(inp_files)} input files')

# Open the files and read the data
inputs = []
for file in inp_files:
    with open(file, 'r') as f:
        inputs.append(f.read().strip())
```

## Day 1: Not Quite Lisp

### Part 1

```python
d1_in = inputs[0]

rb = d1_in.count(')')
print(len(d1_in) - 2 * rb)
```

### Part 2

```python
floor = 0
for i, bracket in enumerate(d1_in):
  if bracket == '(':
    floor += 1
  else:    floor -= 1

  if floor == -1:
    print(i+1)
    break
```

## Day 2: I Was Told There Would Be No Math

### Part 1

```python
d2_in = inputs[1].splitlines()

total = 0

for box in d2_in:
  l, w, h = map(int, box.split('x'))
  total += 2*(l*w + w*h + h*l) + min(l*w, w*h, h*l)
print(total)
```

### Part 2

```python
ribbon = 0

for box in d2_in:
  l, w, h = map(int, box.split('x'))
  ribbon += min(2*(l+w), 2*(w+h), 2*(h+l)) + l*w*h

print(ribbon)
```

## Day 3: Perfectly Spherical Houses in a Vacuum

### part 1

```python
d3_in = inputs[2]


def walk_simulatorLOL(data):
  start = (0,0)
  map = {}

  north = {">":(0,1)}
  south = {"<":(0,-1)}
  east = {"^":(1,0)}
  west = {"v":(-1,0)}

  for i in data:
    if not map.get(start):
      map[start] = 1

    if north.get(i):
      start = (start[0] + north[i][0], start[1] + north[i][1])
    elif south.get(i):
      start = (start[0] + south[i][0], start[1] + south[i][1])
    elif east.get(i):
      start = (start[0] + east[i][0], start[1] + east[i][1])
    elif west.get(i):
      start = (start[0] + west[i][0], start[1] + west[i][1])

  return map

santa_map = walk_simulatorLOL(d3_in)
print(len(santa_map))
```

### Part 2

```python
robo_map = walk_simulatorLOL(d3_in[1::2])
human_map = walk_simulatorLOL(d3_in[0::2])

print(len(robo_map) + len(human_map) - len(set(robo_map.keys()) & set(human_map.keys())))     # getting robo_map ∩ human_map
```
