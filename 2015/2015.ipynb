## Day 1: Not Quite Lisp

### Part 1

```python
file_path = '/home/meek/github/gojo/advent_of_code/2015/data/2015_1.txt'

with open(file_path, 'r') as file:
    inp = file.read().strip()

rb = inp.count(')')
print(len(inp) - 2 * rb)
```

### Part 2

```python
floor = 0
for i, bracket in enumerate(inp):
  if bracket == '(':
    floor += 1
  else:    floor -= 1

  if floor == -1:
    print(i+1)
    break
```
